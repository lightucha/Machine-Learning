## Problem 1

```python
import numpy as np
from random import shuffle, choice
```

```python
n_trials = 50000
newCar = 0

doors = ['car', 'goat', 'goat']

for i in range(n_trials):
    
    gameDoor = doors.copy()
    
    gameShow = gameDoor
    shuffle(gameShow)
    
    firstChoice = np.random.randint(3)
    gameShow.remove(gameShow[firstChoice])
    
    secondChoice = np.random.randint(2)
    gameShow.remove(gameShow[secondChoice])
        
    finalChoice = gameShow
        
    if finalChoice[0] == 'car':
        newCar += 1

    
simulation = newCar / n_trials

staying = simulation * 100
switching = (1 - simulation) * 100

print("Start simulation ... \n")
print("Staying: {:09.6f}%".format(staying))
print("Switching: {:09.6f}%".format(switching))
```

## Problem 2

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
n0 = 200
n1 = 200

sigma = [[19, -4],
         [-4, 1]]

x0 = np.random.multivariate_normal([0.7,0.7], sigma, n0).T        # data in class 0
x1 = np.random.multivariate_normal([-0.5,-0.5], sigma, n1).T      # data in class 0

x0 = np.asmatrix(x0)
x1 = np.asmatrix(x1)
print(x1.shape)
```

```python
mu0 = np.mean(x0, axis=1)
mu1 = np.mean(x1, axis=1)
print("mu0:",mu0.shape)

S0 = 1/(n0-1) * (x0-mu0)*(x0-mu0).T
S1 = 1/(n1-1) * (x1-mu1)*(x1-mu1).T
print("S0:",S0.shape)

w = (n0*S0 + n1*S1).I * (mu0-mu1)
print("w:", w.shape)

plt.figure(figsize = (10, 6))
plt.plot(x0[0,:],x0[1,:],'r.')
plt.plot(x1[0,:],x1[1,:],'b.')

xp = np.arange(-4, 6, 0.1)
yp = w[1,0]/w[0,0]  * xp
plt.plot(xp, yp, 'k')
plt.axis('equal')
plt.ylim([-8, 8])
plt.xlim([-14, 14])
plt.show()
```

```python
y1 = x0.T*w
y2 = x1.T*w
print(y1.shape)

plt.figure(figsize = (10, 6))
plt.hist(y1, 21, color='r', rwidth=0.5)
plt.hist(y2, 21, color='b', rwidth=0.5)
plt.show()
```

## Problem 3

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
n0 = 200
n1 = 200

sigma = [[19, -4],
         [-4, 1]]

x0 = np.random.multivariate_normal([0.7,0.7], sigma, n0)
x1 = np.random.multivariate_normal([-0.5,-0.5], sigma, n1)

x0 = np.asmatrix(x0)
x1 = np.asmatrix(x1)
X = np.vstack((x0, x1))
print(X.shape)

plt.figure(figsize = (10, 6))
#plt.plot(x0[:,0],x0[:,1],'k.')
#plt.plot(x1[:,0],x1[:,1],'k.')
plt.plot(X[:,0],X[:,1],'k.')
plt.ylim([-8, 8])
plt.xlim([-14, 14])
plt.show()
```

```python
m = n0 + n1

S = 1/(m-1)*X.T*X
S = np.asmatrix(S)
 
D, V = np.linalg.eig(S)

idx = np.argsort(-D)
D = D[idx]
V = V[:,idx]

print(D)
print(V)
```

```python
h = V[1,0]/V[0,0]
xp = np.arange(-12, 12, 0.1)
yp = h*xp

fig = plt.figure(figsize=(10, 6))
plt.plot(X[:,0], X[:,1], 'k.')
plt.plot(xp, yp, 'r', linewidth=4.0)
plt.axis('equal')
plt.show()
```

```python
Z = X*V[:,1]

plt.figure(figsize=(10, 6))
plt.hist(Z, 51)
plt.show()
```

## Problem 4

```python
import numpy as np
import random
import matplotlib.pyplot as plt
from scipy import random, linalg
```

```python
A = random.rand(5,5)
A = np.asmatrix(A)
sigma = A.T*A
m = 1000

X = np.random.multivariate_normal([0.,0.,0.,0.,0.],sigma, m)

X = np.asmatrix(X)
print(X.shape)
print(X)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(X[:,0],X[:,1],'k.')
plt.ylim([-6, 6])
plt.xlim([-6, 6])
plt.show()
```

```python
X = X - np.mean(X,axis=0)

S = 1/(m-1)*X.T*X
S = np.asmatrix(S)
 
D, V = np.linalg.eig(S)

idx = np.argsort(-D)
D = D[idx]
V = V[:,idx]

print(D)
print(V)
```

```python
plt.figure(figsize=(10,6))
plt.stem(D/D.sum())
plt.yticks(np.arange(0,1.1,0.1))
plt.title("Variance(%)")
plt.show()
```

We decide 2 or 3 features of data to be remained

```python
# relative magnitutes of the principal components

Z = X*V
xp = np.arange(0,m)/24

plt.figure(figsize=(14, 6))
plt.plot(xp, Z)
plt.yticks([])
plt.show()
```

```python
## projected onto the first principal component
# 6 dim -> 1 dim (dim reduction)
# relative magnitute of the first principal component

Z = X*V[:,0]

plt.figure(figsize=(14, 6))
plt.plot(Z)
plt.yticks([])
plt.show()
```
