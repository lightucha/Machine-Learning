```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
#training data set generation

n0 = 200
n1 = 200

mu = [0, 0]
sigma = [[0.9, -0.4],
         [-0.4, 0.3]]

x0 = np.random.multivariate_normal([2.5,2.5], sigma, n0).T        # data in class 0
x1 = np.random.multivariate_normal([1,1], sigma, n1).T            # data in class 0

x0 = np.asmatrix(x0)
x1 = np.asmatrix(x1)
```

```python
print(x0.shape)
```

```python
mu0 = np.mean(x0, axis = 1)
mu1 = np.mean(x1, axis = 1)
print(mu0.shape); print(mu0)

S0 = 1/(n0-1) * (x0-mu0)*(x0-mu0).T
S1 = 1/(n1-1) * (x1-mu1)*(x1-mu1).T

w = (n0*S0 + n1*S1).I*(mu0 - mu1)
print(w.shape); print(w)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(x0[0,:],x0[1,:],'r.')
plt.plot(x1[0,:],x1[1,:],'b.')

xp = np.arange(-4, 6, 0.1)
yp = w[1,0]/w[0,0]  * xp
plt.plot(xp, yp, 'k')
plt.axis('equal')
plt.ylim([-2, 6])
plt.show()
```

```python
y1 = x0.T*w
y2 = x1.T*w

plt.figure(figsize = (10, 6))
plt.hist(y1, 21, color='r', rwidth=0.5)
plt.hist(y2, 21, color='b', rwidth=0.5)
plt.show()
```

```python
w = np.array([[1],[0]])
w = np.asmatrix(w)
print(w.shape); print(w)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(x0[0,:],x0[1,:],'r.')
plt.plot(x1[0,:],x1[1,:],'b.')
plt.axhline(0, color='k')
plt.axis('equal')
plt.ylim([-2, 6])
plt.show()
```

```python
y1 = x0.T*w
y2 = x1.T*w
print(y1.shape)

plt.figure(figsize = (10, 6))
plt.hist(y1, 21, color='r', rwidth=0.5)
plt.hist(y2, 21, color='b', rwidth=0.5)
plt.show()
```
