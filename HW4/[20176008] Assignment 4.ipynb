```python
#2

import numpy as np
import random
import matplotlib.pyplot as plt



n0 = 200
n1 = 200

sigma = [[19, -4],
         [-4, 1]]

x0 = np.random.multivariate_normal([0.7,0.7], sigma, n0)        # data in class 0
x1 = np.random.multivariate_normal([-0.5,-0.5], sigma, n1)      # data in class 1

x0 = np.asmatrix(x0)
x1 = np.asmatrix(x1)


#make projection

mu0 = np.mean(x0, axis=1)
mu1 = np.mean(x1, axis=1)
S0 =(x0-mu0)*(x0-mu0).T/(n0-1)
S1 =(x1-mu1)*(x1-mu1).T/(n1-1)
w = (n0*S0 + n1*S1).I * (mu0-mu1)


#projection line
xp = np.arange(-14,14,0.1)
yp = w[1,0]/w[0,0]  * xp

plt.figure(figsize = (10, 6))
plt.plot(x0[:,0],x0[:,1],'r.')
plt.plot(x1[:,0],x1[:,1],'b.')
plt.ylim([-8, 8])
plt.xlim([-14, 14])
plt.plot(xp, yp, 'k')
plt.axis('equal')
plt.ylim([-7.5, 12.5])
plt.show()
```
