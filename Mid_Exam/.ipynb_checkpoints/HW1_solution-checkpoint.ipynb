## Problem 5-b

```python
import cvxpy as cvx
import numpy as np

A = np.array([[2, 0],[-1, 1],[0, 2]])
b = np.array([[1, 0, -1]]).T

x = cvx.Variable(2, 1)
objective = cvx.Minimize(cvx.norm(A*x - b, 2))
prob = cvx.Problem(objective)
result = prob.solve()

print(x.value)
```

## Problem 6

```python
import matplotlib.pyplot as plt
import cvxpy as cvx
import numpy as np
import random
```

```python
# Step 1
ones = np.ones([1,100])
zeros = np.zeros([1, 100])

x = np.hstack([ones, zeros])
x_corrupt = x + 0.2*np.random.rand(1, 200)

plt.plot(x[0])
plt.title('Original signal')
plt.show()
```

```python
print(x_corrupt.shape)

plt.plot(x_corrupt[0])
plt.title('Corrupted Signal')
plt.show()
```

```python
# Step 2
beta = 1
x_con = cvx.Variable(1, 200)
objective = cvx.Minimize(cvx.norm(x_con - x_corrupt, 2))
constraints = [cvx.norm(x_con[1:200] - x_con[0:199],1) <= beta]
prob = cvx.Problem(objective, constraints)
result = prob.solve()

x_con = np.asarray(x_con.value)

plt.plot(x_con[0])
plt.title('Reconstructed Signal')
plt.show()
```
