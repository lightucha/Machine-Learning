### Problem 1-2

```python
x = np.linspace(-4.5, 4.5, 10).reshape(-1, 1)
y = np.array([0.9819, 0.7973, 1.9737, 0.1838, 1.3180, -0.8361, -0.6591, -2.4701, -2.8122, -6.2512]).reshape(-1, 1)

xp = np.arange(-4.5, 4.5, 0.01).reshape(-1, 1)

d = [2, 5, 7, 10]
sigma = 1

plt.figure(figsize=(10, 6))

for k in range(4):
    u = np.linspace(-4.5, 4.5, d[k])
    
    A = np.hstack([np.exp(-(x-u[i])**2/(2*sigma**2)) for i in range(d[k])])
    rbfbasis = np.hstack([np.exp(-(xp-u[i])**2/(2*sigma**2)) for i in range(d[k])])
    
    A = np.asmatrix(A)
    rbfbasis = np.asmatrix(rbfbasis)
    
    theta = (A.T*A).I*A.T*y
    yp = rbfbasis*theta
    
    plt.subplot(2, 2, k+1)
    plt.plot(x, y, 'o', markersize=4)
    plt.plot(xp, yp)
    plt.axis([-5, 5, -12, 6])
    plt.title('num RBFs = {}'.format(d[k]), fontsize=10)
    plt.grid(alpha=0.3)

plt.suptitle('Regression', fontsize=15)
plt.show()
```

```python
# Convert to binary image
data0 = data0 > 125
```

```python
def getfeatures(data):
    data0 = data['0']
    data1 = data['1']
    temp0 = []
    temp1 = []
    for i in range(len(data0)):
        for idx,data_i in enumerate([data0,data1]):
            image_i = data_i[i]
            avg_center = image_i[10:20,10:20].reshape(-1,).mean()
            avg_all = image_i.reshape(-1,).mean()
            bias_i = 1
            sample_i = [avg_center,avg_all,bias_i]
            if idx == 0:
                temp0.append(sample_i)
            else:
                temp1.append(sample_i)
                
    features0 = np.array(temp0)
    features1 = np.array(temp1)
    return (features0,features1)
```

```python
features0,features1 = getfeatures(data)

digit0 = plt.scatter(features0[:,0],features0[:,1],c='b',marker=".",)
digit1 = plt.scatter(features1[:,0],features1[:,1],c='r',marker=".")

plt.xlabel("feature0")
plt.ylabel("feature1")
plt.legend([digit0, digit1], ['digit0', 'digit1'])
plt.grid(alpha=0.3)
plt.title("Feature Space")

plt.show()
```

```python
X = np.concatenate((features0,features1))
Y = np.array([-1 if i < 1000 else 1 for i in range(2000)])
W = np.zeros([3,1]).reshape(-1,)
idx = list(range(2000))
wrongPredArr = []
#n_iter = int(Y.shape[0])
n_iter = 100
for _ in range(n_iter):
    wrongPred  = 0
    for _ in range(n_iter):
        i = np.random.choice(idx)
        yhat = np.sign(np.dot(X[i],W))
        if int(yhat) != int(Y[i]):
            W += Y[i]*X[i]
            wrongPred += 1
    wrongPredArr.append(wrongPred)
    
plt.plot(wrongPredArr)
plt.legend(loc = 4, fontsize = 15)
plt.show()
```

```python
x1p = np.linspace(0,int(max(X[:,0])+5),100).reshape(-1,)
x2p = -(W[0]*x1p)/W[1] - W[2]/W[1]
features0,features1 = getfeatures(data)
digit0 = plt.scatter(features0[:,0],features0[:,1],c='b',marker=".",label='digit0')
digit1 = plt.scatter(features1[:,0],features1[:,1],c='r',marker=".",label='digit1')
hyperP = plt.plot(x1p, x2p, c='k', label='perceptron')
plt.xlabel("feature0")
plt.ylabel("feature1")
plt.legend(loc = 4, fontsize = 10)
plt.title("Feature Space")
plt.show()
```
