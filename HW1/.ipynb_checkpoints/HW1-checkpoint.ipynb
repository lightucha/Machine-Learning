## Problem 6

step 1.

```python
data = np.zeros(200)
data[0:100] = 1

plt.plot(data)
plt.show()
```

```python
noise = np.random.normal(0, 0.1, 200)
noise

corrupted_signal = data + noise

plt.plot(corrupted_signal)
plt.show()
```

step 2.

beta가 작은 값을 가질 때

```python
beta = 1

x_con = cvx.Variable(200)
objective = cvx.Minimize(cvx.norm(x_con - corrupted_signal, 2))
constraints = [cvx.norm(x_con[1:200] - x_con[0:199],1) <= beta]
prob = cvx.Problem(objective, constraints)
result = prob.solve()

plt.plot(x_con.value)
plt.show()

print(x_con.value); print(result)
```

beta가 큰 값을 가질 때

```python
beta = 100

x_con = cvx.Variable(200)
objective = cvx.Minimize(cvx.norm(x_con - corrupted_signal, 2))
constraints = [cvx.norm(x_con[1:200] - x_con[0:199],1) <= beta]
prob = cvx.Problem(objective, constraints)
result = prob.solve()

plt.plot(x_con.value)
plt.show()

print(x_con.value); print(result)
```

In addition, by changing the beta, explain how beta affects in an optimization process.

beta 값이 커질수록 모델은 overfitting되는 경향이 생긴다. beta 값이 커진다는 것은 계수 추정치들이 커진다는 의미이며, 최소제곱 솔루션(least-square)이 되어 감을 뜻한다. 결국, 해당 optimization 문제는 최상의 부분집합 선택과 밀접하게 관련되어 있음을 알 수 있다.
